##  1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize = (12,8))
plt.hist(df['price'], bins = 40)
plt.title("Стоимость недвижимости")
plt.xlabel('Цена')
plt.ylabel("количество домов")
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(data = df['price'])
plt.title("Стоимость недвижимости")
plt.ylabel("Стоимость")
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(data = df['price'], showfliers=False)
plt.title("Стоимость недвижимости")
plt.ylabel("Стоимость")
```

#### Выводы 1.2
1. Распределение стоимости домов неравномерно. 
2. Основное количество домов приходится на ценовой диапзон от 350 тысяч до 650 тысяч. 
3. Дома более высокой ценовой категории представлены в очень широком диапазоне, однако дома ценой свыше 2 млн представлены единичными позициями. 

### 1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['sqft_living'], bins=30)

plt.title('Распределение жилой площади')
plt.xlabel('Площадь')
plt.ylabel('Кол-во домов');
```

```python
plt.figure(figsize = (6,4))
sns.boxplot(data = df['sqft_living'])
plt.title("Распределение жилой площади")
plt.ylabel("Площадь")
```

#### Выводы 1.3
1. Распределение площади домов по своей структуре похоже на распределение стоимости: дома с большой площадью представлены в меньшем количестве. Однако, если не учитывать совсем уж большие дома, то распределение больше похоже на нормальное, хотя и со сдвигом в сторону меньшей площади
2. Основное количество домов приходится на диапзон площади от  ~1000 до ~2500 м кв.
3. Дома с площадью более 5000 м кв представлены в очень широком диапазоне, однако дома с площадью свыше 8000 м кв  - единичными позициями. 
4. Можно предположить наличие сильной корреляции между стоимостью и площадью домов. Однако для домов с небольшой площадью вероятно на стоимость будут сильно влиять и другие факторы. 

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 4))

sns.histplot(df['yr_built'], bins=60)

plt.title('Распределение года постойки')
plt.xlabel('Год')
plt.ylabel('Кол-во домов')
```

```python
plt.figure(figsize = (6,4))
sns.boxplot(data = df['yr_built'])
plt.title("Распределение годов постройки")
plt.ylabel("годы")
```

#### Выводы 1.4
1. В целом имеется тенденция к увеличению домов более поздних готов посторойки, что вероятно можно объяснить как сносом более старых домов, так и в целом увеличением объемов строительства
2. Кроме того, можно выявить максимумы и минимумы. Например, низкое количество домов 1930-х годов постройки по сравнению с 1920 можно объяснить Великой депрессией, а минимум в 1940 - вступлением США во вторую мировую войну. Так же можно отметить пики строительства в конце 70 и конце 2000х, которые можно объяснить экономическим ростом.
3. Большая часть домов была построена в промежутке от 1950 до 2000 года.

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()
data.index = ['no', 'yes']
data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(data, autopct='%1.2f%%')

plt.title('Доля домов с видом на набережную')
plt.legend(data.index);
```

#### Выводы:
Доля домов с видом на набержную составляет менее одного процента ~0.75%.

### 2.2 Изучите распределение этажей домов

```python
data2 = df['floors'].value_counts()
data2.head()
```

```python
plt.figure(figsize=(8, 6))    
plt.title('Распределение этажей домов')
sns.barplot(x=data2.index, y=data2.values)
plt.xlabel('этажность')
plt.ylabel('Кол-во домов')    
plt.show()
```

```python
plt.figure(figsize=(6, 4))

plt.pie(data2, autopct='%1.2f%%')

plt.title('Доля домов с различной этажностью')
plt.legend(data2.index);
```

#### Выводы 2.2
